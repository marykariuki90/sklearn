encoding techniques

```python
#label encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv('IRIS.csv')
df
```

```python
#encode labels in species column
#used in ordinal variables that require rearrangement based on ranking
encode=LabelEncoder()
df['species']=encode.fit_transform(df['species'])
df['species'].unique()
```

# one_hot encoding using pandas

```python
#one-hot encoding
#its is used in norminal variables which are simply unordered
one_hot=pd.get_dummies(df['species'])
print(one_hot.to_string())
```

# one_hot encoding using sklearn

```python
#one _hot encoding it eliminates one category,
#that is total categories minus 1
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder()
x=df['species']=encode.fit_transform(df[['species']]).toarray()
print(x)
```
