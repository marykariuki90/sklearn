# rescaled data(MinMaxScaler())
it returns all values in range between 0 and 1

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
df=pd.read_csv('IRIS.csv')
df
```

```python
#converting from csv format to dataframe for slicing purposes
arr=df.values
arr
```

```python
feature_range=['sepal_length','sepal_width','petal_length','petal_width']
```

```python
#slicing the data
x=arr[:,0:4]
y=arr[:,4]
scaling=MinMaxScaler(feature_range=(0,1))
rescale_x=scaling.fit_transform(x)
numpy.set_printoptions(precision=3)
print(rescale_x[0:5,:])
```

# binarize data(binarize())
it is where all the values above the threshold are marked as 1 and below the threshold are marked as 0

```python
from sklearn.preprocessing import Binarizer
X=arr[:,0:4]
y=arr[:,4]
scale_x=Binarizer(threshold=0.5).fit_transform(X)
numpy.set_printoptions(precision=3)
print(scale_x[0:20,:])
```

# standardize data(standardscaler())
it uses mean of 0 and standard deviation of 1


```python
from sklearn.preprocessing import StandardScaler
X=arr[:,0:4]
y=arr[:,4]
scaled_x=StandardScaler().fit_transform(X)
numpy.set_printoptions(precision=3)
print(scaled_x[0:20,:])
```
